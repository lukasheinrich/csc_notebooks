# Requirements

The excercises introduce basic concepts in Machine Learning and don't require a whole lot of special libraries.

The most important onces are `scipy`, `numpy`, `matplotlib` and `torch` (for deep learning)

Execute the following cells to see if they work

```python
import matplotlib.pyplot as plt
import scipy.stats as sps
import numpy as np
```

For example we could plot a Gaussian density

```python
xi = np.linspace(-10,10)
yi = sps.norm(1,3).pdf(xi)
plt.plot(xi,yi)
plt.gcf().set_size_inches(3,3)
```

We can also test that `torch` works

```python
import torch
```

To first order `torch` is a library like `numpy` for array computation, but comes with additional
tools for 

* autoamtic differentiation
* building and training neural networks

```python
def f(x):
    return torch.pow(x,3)
```

```python
xi = torch.linspace(-5,5,101)
yi = f(xi)
plt.plot(xi,yi)
plt.gcf().set_size_inches(3,3)
```
